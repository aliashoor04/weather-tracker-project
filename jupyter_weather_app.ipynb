# Ali Part(menu one):
**your function should:**
- make sure the csv file exists
- if it does not exist, create it with headers

**the user input should be:**
- date MM-DD-YYYY format
- temperature in celsius
- condition sunny, cloudy, rainy, snowy
- humidity in presentage
- wind speed in km/h- Save the observation into the CSV

```python
# importing the needed modules
import csv
from datetime import datetime
from collections import Counter

# defining the first menu fuction
def record_observation():
    """Adding a new weather observation into the weather csv file""" # the docstring od the function

    try: # try to open the file if it exists 
        with open('weather.csv', 'r', newline='') as file:
            reader = csv.DictReader(file)
            for row in reader: # convert values to float since csv stores everything as text
                row['temperature'] = float(row['temperature'])
                row['humidity'] = float(row['humidity'])
                row['wind_speed'] = float(row['wind_speed'])
        print('\nweather.csv exists.')
    except FileNotFoundError: # create the file if it does not exist
        with open('weather.csv', 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['date', 'temperature', 'condition', 'humidity', 'wind_speed']) # the column header of our csv
            print('\nnew weather.csv created')

    print("\n=== Record a New Weather Observation ===")

    while True: # enter the date 
        observation_date = input('Enter date (MM-DD-YYYY): ')
        try:
            datetime.strptime(observation_date, "%m-%d-%Y")
            break # to exit the loop
        except ValueError: # check the format  
            print("Invalid date format. Please use MM-DD-YYYY.")

    while True: # enter the temp 
        try:
            temperature = float(input("Enter temperature in degrees Celsius: "))
            break # to exit the loop
        except ValueError: ## check the format 
            print("Please enter a valid number.")

    valid_conditions = ['sunny', 'cloudy', 'rainy', 'snowy'] # list of allowed condition

    print("\nAvailable conditions: Sunny, Cloudy, Rainy, Snowy")

    while True:
        condition = input("Enter weather condition: ").lower() # convert input to lowercase to compare with our list
    
        if condition in valid_conditions: # forcing the user to choose from a list
            condition = condition.capitalize()  # makes it capital latter
            break
        else:
            print("Invalid input. Please choose from: Sunny, Cloudy, Rainy, Snowy.")    
            

    while True: # enter the humidity 
        try:
            humidity = float(input("Enter humidity percentage: "))
            if 0 <= humidity <= 100: # must be in percentage
                break # to exit the loop
            print("Please enter a value between 0 and 100.")
        except ValueError: # check the format 
            print("Please enter a valid number.")

    while True: # enter the speed 
        try:
            wind_speed = float(input("Enter wind speed in km/h: "))
            if wind_speed >= 0: # must be posituve
                break # to exit the loop
            print("Please enter a non-negative number.")
        except ValueError: # check the format 
            print("Please enter a valid number.")

    # creating a dictionary for the new observation
    new_observation = {
        'date': observation_date,
        'temperature': temperature,
        'condition': condition,
        'humidity': humidity,
        'wind_speed': wind_speed
    }

    try: # add or append  the new observation 
        with open('weather.csv', 'a', newline='') as file:
            fn = ['date', 'temperature', 'condition', 'humidity', 'wind_speed']
            writer = csv.DictWriter(file, fieldnames=fn)
            writer.writerow(new_observation)
            print(f"\nWeather observation for {observation_date} added successfully!")
    except Exception as e:
        print(f'Error saving new weather observation: {e}')
```

# Muhammed Part(menu two):
**your function should:**

- open weather.csv using DictReader
- store temperatures and conditions in lists
- calculate min, max, average, and most common condition
- if data exists  **print statistics**
- if empty  **print "No weather data found"**
- handle file error using try-except

```python
def view_statistics():
    try:
        with open('weather.csv', 'r', newline='') as file:
            reader = csv.DictReader(file)
            temperature = []
            condition=[]

            for row in reader:
                temperature.append(float(row['temperature']))
                condition.append(row['condition'])
                
            min_temp=min(temperature)
            max_temp=max(temperature)
            avg_temp=sum(temperature) / len(temperature)
            most_condition = Counter(condition).most_common(1)[0][0]
            if temperature:
                print('\n---Weather Statistic---')
                print(f'The Minimum Temperature is {min_temp} C')
                print(f'The Maximum Temperature is {max_temp} C')
                print(f'The Average Temperature is {avg_temp:.2f} C')
                print(f'The Most Common Condition is {most_condition}.')
            else:
                print('No weather data found.')

    except FileNotFoundError:
        print('error: weather.csv file not found')

        
```

```python
view_statistics()
```

```python
import csv
def search_by_date():
    try:
        search_date = input("Enter date to search for (MM-DD-YYYY): ")

        with open('weather.csv', 'r', newline='') as file:
            reader = csv.DictReader(file)
            found = False

            for row in reader:
                if row['date'] == search_date:
                    if not found:
                        print(f"\n--- Observations for {search_date} ---")
                        found = True

                    print(f"Temperature: {row['temperature']} C")
                    print(f"Condition: {row['condition']}")
                    print(f"Humidity: {row['humidity']}%")
                    print(f"Wind Speed: {row['wind_speed']} km/h")
                    print("-" * 30)

            if not found:
                print("No observations found for that date.")

    except FileNotFoundError:
        print("error: weather.csv file not found")
```

```python
search_by_date()
```

```python
def view_all_observations():
    print("\n      --- All Weather Observations ---\n")

    try:
        with open('weather.csv', 'r', newline='') as file:
            reader = csv.DictReader(file)

            # top border
            print("+------------+------------+--------------+----------------+------------------+")

            # header
            print("| Date       | Temp (C)   | Condition    | Humidity (%)   | Wind Speed (km/h)|")

            # separator
            print("+------------+------------+--------------+----------------+------------------+")

            for row in reader:
                print(f"| {row['date']:<10} | {row['temperature']:<10} | {row['condition']:<12} | {row['humidity']:<14} | {row['wind_speed']:<16} |")

            # bottom border
            print("+------------+------------+--------------+----------------+------------------+")

    except FileNotFoundError:
        print("Error: weather.csv file not found.")
```

```python
view_all_observations()
```

# Main Code 


```python
def main():
    while True:
        print("\n----The Weather Tracker App----")
        print("1. Record observation")
        print("2. View statistics")
        print("3. Search by date")
        print("4. View all observations")   
        print("5. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            record_observation()
        elif choice == "2":
            view_statistics()
        elif choice == "3":
            search_by_date()
        elif choice == "4":
            view_all_observations()   
        elif choice == "5":
            print("Thank you for using The Weather Tracker App. Now go enjoy the real weather :)")
            break
        else:
            print("Invalid choice")
```

```python
main()
```
